```python
import requests
from bs4 import BeautifulSoup
```

```python
web=requests.get("https://www.tutorialsfreak.com/")
```

# Parse the webpage's raw HTML into a BeautifulSoup object for structured tag extraction

```python
soup=BeautifulSoup(web.content,"html.parser")
```

# Prints the BeautifulSoup object structure in a readable (indented) HTML format

```python
print(soup.prettify)
```

# Loops through all <a> tags and prints each link's URL from the 'href' attribute.

```python
for i in soup.find_all("a"):
    print(i.get("href"))
```

EXTRACTING IMG


# Finds and stores all <img> tags from the webpage in a list for image extraction.

```python
img=soup.find_all("img")
img
```

```python
for i in img:
    print(i.get("src"))
    print(type(i.get("src")))
```

```python
for i in img:
    print(i.get("alt"))
    print(type(i.get("src")))
```
