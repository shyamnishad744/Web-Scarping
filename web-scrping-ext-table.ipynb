```python
import requests
import pandas as pd
from bs4 import BeautifulSoup 
```

```python
url="https://ticker.finology.in/"
```

```python
r=requests.get(url)
print(r)
```

```python
soup=BeautifulSoup(r.text,"lxml")
```

```python
table=soup.find("table",class_="table table-sm table-hover screenertable")
#print(table)
headers=table.find_all("th")
print(headers)
```

```python
titles=[]               # Create an empty list to store table headers
for i in headers:        # Loop through each header element found earlier
    title=i.text         # Extract the text content from each header tag
    titles.append(title)   # Add the extracted text to the 'titles' list
    
print(titles)              # Print the list of all extracted table headers
```

```python
df=pd.DataFrame(columns=titles)   # Create an empty DataFrame with column names from the 'titles' list
print(df)                         # Display the empty DataFrame to check the column headers
```

```python
rows=table.find_all("tr")    # Find all table row (<tr>) elements from the HTML table

for i in rows[1:]:             # Loop through each row except the first (header row)
    data=i.find_all("td")          # Extract all table data (<td>) cells from the current row
    #print(data)               # (Optional) Check the extracted data cells

    rows=[tr.text for tr in data]   # Create a list of text values from each <td> cell
    #print(rows)                       # (Optional) View the cleaned row data

    l=len(df)                       # Get the current number of rows in the DataFrame
    df.loc[l]=rows                    # Insert the new row data into the next available index

    print(df)                    # Print the DataFrame after adding each new row 
```
